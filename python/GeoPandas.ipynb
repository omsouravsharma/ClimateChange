```python
chicago.plot(
    column = "POP2010",
#     ax = ax,
    legend = True,
    legend_kwds = {"label":"Population in 2010", "orientation":"horizontal"},
);
```

```python
# Populate population estimate with correct legend

from mpl_toolkits.axes_grid1 import make_axes_locatable
fig, ax = plt.subplots(1,1)
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size = "5%", pad = 0.1)

chicago.plot(
    column = "POP2010", 
    ax =ax ,
    legend = True, 
    cax = cax, 
    legend_kwds = {"label":"Population in 2010", "orientation":"horizontal"}
);
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
fig, ax = plt.subplots(1,1)
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size = "5%", pad = 0.1)

chicago.plot(
    column = "POP2010", 
    ax =ax ,
    legend = True, 
    cax = cax, 
    cmap = 'OrRd',
    legend_kwds = {"label":"Population in 2010", "orientation":"horizontal"}
);
```
